```python
from django.db.models import Q
```

```python
qs = Post.objects.all()
qs = qs.filter(id__gte=2, message__icontains=query)
print(qs.query)
```

```python
qs = Post.objects.all()
qs = qs.filter(Q(id__gte=2), Q(message__icontains=query))
print(qs.query)
```

```python
qs = Post.objects.all()
qs = qs.filter(id__gte=2, message__icontains=query)
qs = qs.filter(Q(id__gte=2) & Q(message__icontains=query))
print(qs.query)
```

```python
qs = Post.objects.all()
qs = qs.filter(Q(id__gte=2) | Q(message__icontains=query))
print(qs.query)
```

```python
qs = Post.objects.all()

cond = Q(id__gte=2) | Q(message__icontains=query)
# cond &= Q(...)

qs = qs.filter(cond)
print(qs.query)
```
